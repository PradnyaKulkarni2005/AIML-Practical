```python
# Code - 12
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Student_performance_data _.csv")
print(df.head())
```

```python
X=df['StudyTimeWeekly']
y=df['GPA']
```

```python
# NORMALIZATION
X=(X-np.mean(X))/np.std(X)
y=(y-np.mean(y))/np.std(y)

# initialization
m=0
b=0
epoch=1000
learning_rate=0.1
n=len(X)
# loop through epoch
for i in range (epoch):
    y_pred=m*X+b
    error=y_pred-y

    # gradient descents
    dm=(2/n)*np.dot(X,error)
    db=(2/n)*np.sum(error)
    m-=learning_rate*dm
    b-=learning_rate*db

y_pred=m*X+b
# calculate the MEAN squared error and r2 score
MSE=np.mean((y-y_pred)**2)
R2=1-((np.mean(y-y_pred)**2))/(np.mean(y - np.mean(y)**2 ))
# print result
print("Slope-",m)
print("Intercept- ",b)
print("MSE - ",MSE)
print("R2 score",R2)

```
